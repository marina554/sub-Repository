```python
import requests
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# あなたのAPIキーを入力
api_key = "（ここにあなたのAPIキー）"

# 調べたい都市リスト
cities = ["Tokyo", "Osaka", "Sapporo", "Nagoya", "Fukuoka"]

# 空のリストを用意
weather_list = []

# 各都市のデータを取得
for city in cities:
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&lang=ja&units=metric"
    response = requests.get(url)
    data = response.json()

    if response.status_code == 200:
        info = {
            "都市": data["name"],
            "天気": data["weather"][0]["description"],
            "気温(℃)": data["main"]["temp"],
            "体感温度(℃)": data["main"]["feels_like"],
            "湿度(%)": data["main"]["humidity"],
            "風速(m/s)": data["wind"]["speed"],
            "気圧(hPa)": data["main"]["pressure"]
        }
        weather_list.append(info)
    else:
        print(f"{city} のデータ取得に失敗しました")

# DataFrameに変換して並べ替え
df = pd.DataFrame(weather_list)
df = df.sort_values(by="気温(℃)", ascending=False)

# Excel出力
excel_path = "/content/weather_multi.xlsx"
df.to_excel(excel_path, index=False)

# Excelを開いてセルの色をつける
wb = load_workbook(excel_path)
ws = wb.active

# 赤の塗りつぶし設定
red_fill = PatternFill(start_color="FF9999", end_color="FF9999", fill_type="solid")

# 見出しを除く2行目以降を対象にする
for row in range(2, ws.max_row + 1):
    temp = ws[f"C{row}"].value  # 「気温(℃)」列（C列）
    if temp >= 25:
        ws[f"C{row}"].fill = red_fill

wb.save(excel_path)
print("気温が25℃以上のセルを赤く塗りました:", excel_path)
```
