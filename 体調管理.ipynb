```python
import csv
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import os

# === 1. ユーザー入力 ===
score_input = input("今日の体調は何点ですか？（0～100点で入力してください）：")

try:
    score = int(score_input)
    if not (0 <= score <= 100):
        print("0～100の範囲で入力してください。")
        exit()

except ValueError:
    print("数値を入力してください。")
    exit()

# === 2. メッセージを表示 ===
if score <= 50:
    print("今日は無理をしないでください。")
elif score <= 70:
    print("普段どおりいきましょう。")
else:
    print("いいですね！その調子です！")

# === 3. 日付と一緒にCSVに保存 ===
date = datetime.date.today()
filename = "health_log.csv"
file_exists = os.path.exists(filename)

with open(filename, "a", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    # 初回のみヘッダーを書く
    if not file_exists:
        writer.writerow(["日付", "体調スコア"])
    writer.writerow([date, score])

print(f"{date} の体調スコア（{score}点）を記録しました。")

```
